### 4. Model building

```python
X = cancer_data_df.drop(labels=['target'],axis=1)
y = cancer_data_df[['target']]
```

```python
X.shape, y.shape
```

### Data Preprocessing for continous data
* StandadScalar - mean = 0 sd = 1 > standardization
* minmax Scalar - 0 - 1        > Normalization
* robust Scalar

```python
X.mean()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
std_scaler = StandardScaler()
scaled_X = std_scaler.fit_transform(X)
```

```python
scaled_X = pd.DataFrame(data = scaled_X, columns=cancer_data.feature_names)
scaled_X
```

```python
scaled_X.mean()
```

```python
scaled_X.std()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, shuffle= True,  random_state=12, stratify=y)
```

```python
y_test.value_counts()
```

```python
plt.pie(x = y_train.value_counts(),autopct='%2f',)
plt.show()
```

```python
plt.pie(x = y_train.value_counts(),autopct='%2f',)
plt.show()
```

### model building

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
%time
dt_model = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, class_weight={0:1,1:2})#, class_weight={0:1,1:2})
dt_model.fit(X_train,y_train)

# ad_model = AdaBoostClassifier()
# ad_model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator=dt_model,param_grid={'criterion':['gini','entropy'],
    'max_depth' : np.arange(20)})
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```

training accuracy = 96
testing accuracy = 87

### 6. Model Testing 7. Model Evaluation

##### Training data

```python
y_pred_train = dt_model.predict(X_train)

from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
print('accuracy', accuracy_score(y_train, y_pred_train))
print(confusion_matrix(y_train, y_pred_train))
print('classification report', classification_report(y_train, y_pred_train))
```

```python
y_pred_train = dt_model.predict(X_train)

from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
print('accuracy', accuracy_score(y_train, y_pred_train))
print(confusion_matrix(y_train, y_pred_train))
print('classification report', classification_report(y_train, y_pred_train))
```

```python
y_pred_test = dt_model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
print('accuracy', accuracy_score(y_test, y_pred_test))
print(confusion_matrix(y_test, y_pred_test))
print('classification report', classification_report(y_test, y_pred_test))
```
